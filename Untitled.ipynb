```python
import yt_dlp
import json
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# 1. SCRAPE TRANSCRIPT + METADATA
url = "https://www.youtube.com/watch?v=BxV14h0kfzM"  # sample video

ydl_opts = {
    "skip_download": True,
    "writeinfojson": True,
    "writesubtitles": True,
    "subtitleslangs": ["en"]
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

print("Done: Transcript + metadata saved!")
```
