```python
import yt_dlp

url = "https://www.youtube.com/watch?v=5wYq8spIBpk"

ydl = yt_dlp.YoutubeDL({"skip_download": True})
info = ydl.extract_info(url, download=False)

print("Manual subtitles:", info.get("subtitles", {}).keys())
print("Auto subtitles:", info.get("automatic_captions", {}).keys())
```

```python
import yt_dlp

url = "https://www.youtube.com/watch?v=5wYq8spIBpk"

options = {
    "format": "bestaudio/best",
    "outtmpl": "audio.%(ext)s"
}

with yt_dlp.YoutubeDL(options) as ydl:
    ydl.download([url])

print("Audio downloaded!")
```

```python
import whisper

model = whisper.load_model("medium")  # if too slow, use "small"
result = model.transcribe("audio.webm")

text = result["text"]
text[:500]
```

```python
with open("transcript.txt", "w", encoding="utf-8") as f:
    f.write(text)
```
